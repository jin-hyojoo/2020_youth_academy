```python
import pandas as pd
popu = pd.read_csv('book/population.csv', index_col='State')
popu.head()
```

```python
alco2009 = pd.read_csv('book/niaaa-report2009.csv', index_col='State')
alco2009.head()
```

```python
# merger 후 인덱싱  (== 이너조인 형식)
df=pd.merge(alco2009.reset_index(), popu.reset_index()).set_index('State')
df.head(3)
```

```python
# 아우터 조인 형식
df = pd.merge(alco2009, popu, left_index= True, right_index=True)
df.head(3)
```

```python
#population 값 기준으로 인덱스 값 같은 데이터만 갖고오기
popu.join(alco2009).tail(10)
```

```python
# concat을 조인처럼 사용하기 (기준축이 있기 때문에 가능) => 컬럼 기준으로 
pd.concat([alco2009, popu], axis=1).tail()
```

```python
# 캐나다 미국의 population 합치기 
popu_ca = pd.read_csv('book/demo02a-eng.csv', index_col='State')
popu_ca.head(3)
```

```python
# concat은 위 아래로 붙음 (키 값을 주어 미국, 캐나다 값 구분)
pop_na = pd.concat([popu, popu_ca], keys=["US", "CA"])
pop_na.head()
```
