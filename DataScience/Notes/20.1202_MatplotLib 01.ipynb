```python
# subgrid 이용해 영역 합치거나 나누기 (영역분배)
# 영역이 겹치면 덮어쓰기 때문에 원하는 모양이 나오지 않음 주의하기
fig = plt.figure(0, figsize=(12,9))
plt.subplot2grid((3,3),(0,0))

# colspan 합치기, 자기 자신포함해 오른쪽으로 합침
plt.subplot2grid((3,3),(0,1), colspan=2)

plt.show()
```

```python
iris = pd.read_csv('../Example Codes/Lec03 Matplotlib/data/iris_dataset.csv')
iris.species.unique()
```

```python
fig = plt.figure(0, figsize=(9,4)) # size 조절
plt.subplot(1,2,1)
plt.scatter(iris.petal_length, iris.petal_width)
plt.subplot(1,2,2)
plt.scatter(iris.sepal_length, iris.sepal_width)
```

```python
iris.head()
```

```python
def check(x):
    if x == 'setosa':
        return 0
    elif x == 'versicolor':
        return 1
    else:
        return 2
    
# map엔 값이 하나씩 들어감 (하나씩 비교해가며 리턴) 
iris.species.map(check)
```

```python
# 위에 함수 지정해서 map()한 코드를 한 줄로
iris['species_color'] = iris.species.map({'setosa':0, 'versicolor':1, 'virginica':2}) # 구분값 담을 뉴 컬럼 추가
iris.tail(3)
```

```python
fig = plt.figure(0, figsize=(9,4)) # size 조절

# scatter 1
plt.subplot(1,2,1)
plt.scatter(iris.petal_length, iris.petal_width, c = iris.species_color)


# scatter 2
plt.subplot(1,2,2)
plt.scatter(iris.sepal_length, iris.sepal_width, c = iris.species_color)
```

```python
fig = plt.figure(0, figsize=(9,4)) # size 조절

# scatter 1
plt.subplot(1,2,1)
plt.scatter(iris.petal_length, iris.petal_width, c = iris.species_color, s=50*iris.petal_length*iris.petal_width, alpha = 0.3)


# scatter 2
plt.subplot(1,2,2)
plt.scatter(iris.sepal_length, iris.sepal_width, c = iris.species_color, s=50*iris.petal_length*iris.petal_width, alpha = 0.3)
```

```python
month_num = np.arange(1,13).tolist() #1-12월 생성
units_sold = np.random.randint(200,800,12) #월별 판매량
units_sold
```

```python
plt.bar(month_num,units_sold)
plt.plot(month_num,units_sold)
```

```python
import calendar 

# month_num값을 calendar에 있는 값으로 변경
# rotation 옵션을 통해 글씨가 겹쳐 보이는 걸 조정
plt.xticks(month_num, calendar.month_name[1:13], rotation=35) 
bars = plt.bar(month_num,units_sold)

bar0 = bars[0]

plt.text(7, 600, 600) # 화면에서 문자를 찍을 위치
plt.text(bars[0].get_x(), bars[0].get_height(), bars[0].get_height())
plt.show()
```

```python
# 그래프 바에 값 표시
ax = plt.axes()
plt.xticks(month_num, calendar.month_name[1:13], rotation=30)


bars = ax.bar(month_num,units_sold)
for graphe in bars:
    height = graphe.get_height()
    ax.text(graphe.get_x()+graphe.get_width()/2, 
            height*1.002, height,ha="center", va="bottom")
plt.show()
```

```python
# 가로 막대 그래프
plt.yticks(month_num, calendar.month_name[1:13], rotation=30)
bars = plt.barh(month_num,units_sold)
bar0 = bars[0]
#plt.text(bar0.get_width(), bar0.get_y(), bar0.get_width(), va='bottom')

# 막대기 하나에 해당되는 값 출력
for grap in bars:
    width = grap.get_width()
    plt.text(width+10, grap.get_y(), width, va="bottom")

# 그래프 넓히기    
plt.xlim(0,1000)
```
