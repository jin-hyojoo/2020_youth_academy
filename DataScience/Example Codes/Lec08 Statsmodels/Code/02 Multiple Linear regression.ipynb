```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
data = pd.read_csv('data/Multiple linear regression.csv')
```

```python
data.head()
```

```python
#y = data['GPA']
y = data.GPA
orginal_x = data[['SAT', 'Rand 1,2,3']]
```

```python
x = sm.add_constant(orginal_x)
```

```python
x.head()
```

```python
results = sm.OLS(y,x).fit()
```

```python
results.summary()
```
