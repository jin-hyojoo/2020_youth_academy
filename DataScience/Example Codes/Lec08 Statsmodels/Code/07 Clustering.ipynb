```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('data/Country clusters.csv')
```

```python
data
```

```python
plt.scatter(data.Longitude, data.Latitude)
plt.xlim(-180, 180)
plt.ylim(-90,90)
```

```python
x = data.iloc[:, 1:3]
x.head(2)
```

```python
kmeans = KMeans(3)
```

```python
kmeans.fit(x)
```

```python
cluster = kmeans.fit_predict(x)
cluster
```

```python
data_with_cluster = data.copy()
data_with_cluster['cluster'] = cluster
data_with_cluster
```

```python
plt.scatter(data_with_cluster.Longitude, data_with_cluster.Latitude,
           c =data_with_cluster.cluster, cmap='rainbow' )
plt.xlim(-180, 180)
plt.ylim(-90,90);
```

```python
data_map = data.copy()
data_map['Language'] = data_map['Language'].map(
    {'English':0, 'French':1,'German':2})
```

```python
kmeans = KMeans(2)
kmeans.fit(data_map.iloc[:, 1:4])
data_map['cluster'] = kmeans.fit_predict(data_map.iloc[:, 1:4])
data_map
```

```python
plt.scatter(data_map.Longitude, data_map.Latitude,
           c =data_map.cluster, cmap='rainbow' )
plt.xlim(-180, 180)
plt.ylim(-90,90);
```

```python
wcss=[]
for i in range(1,7):
    kmeans= KMeans(i)
    kmeans.fit(x)
    wcss_iter= kmeans.inertia_
    wcss.append(wcss_iter)
```

```python
number_cluseter = range(1,7)
plt.plot(number_cluseter,wcss)
plt.title("The Elbow Method")
plt.xlabel("Number of Cluster")
plt.ylabel("Within-Cluster Sum of Square");
```

```python
data = pd.read_csv('data/iris-dataset.csv')
data.head(2)
```

```python
plt.scatter(data.sepal_length, data.sepal_width)
```

```python
x = data.copy()
kmeans = KMeans(2)
kmeans.fit(x)
result = data.copy()
result['cluster'] = kmeans.fit_predict(x)
plt.scatter(result.sepal_length, result.sepal_width,
            c=result.cluster, cmap='rainbow')
```
