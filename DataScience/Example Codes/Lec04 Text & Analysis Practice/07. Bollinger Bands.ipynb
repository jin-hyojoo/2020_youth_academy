```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/time/apple_stock.csv',index_col='Date', parse_dates=True)
```

```python
df['Close: 20day Mean'] = df['Close'].rolling(20).mean()
```

```python
df['Upper'] = df['Close: 20day Mean']+2*df['Close'].rolling(20).std()
```

```python
df['Lower'] = df['Close: 20day Mean']-2*df['Close'].rolling(20).std()
```

```python
df.tail()
```

```python
df['2018':][['Close', 'Close: 20day Mean','Upper', 'Lower']].plot(figsize=(15,7));
```
