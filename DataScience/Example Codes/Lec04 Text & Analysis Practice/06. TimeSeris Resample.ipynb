```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/time/apple_stock.csv',index_col='Date', parse_dates=True)
```

```python
df.head(4)
```

```python
# sum, mean, median,min, max, std, var
df.groupby([df.index.year, df.index.month]).agg('median')
```

```python
#https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#dateoffset-objects
```

```python
df.resample(rule='A').mean()
```

```python
def first_day(sample):
    return sample[0]
```

```python
df.resample(rule='A').apply(first_day)
```

```python
df[['Close']].resample('A').mean().plot(kind='bar')
plt.title('Yearly mean close price of apple');
```

```python
df[['Open']]['2015':].resample('M').max().plot(kind='bar', figsize=(15,7))
plt.title("Monthly Max Opening Price for Apple since 2015");
```
