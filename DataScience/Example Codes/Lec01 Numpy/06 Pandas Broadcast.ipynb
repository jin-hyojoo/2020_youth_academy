```python
import pandas as pd
```

```python
scientists = pd.read_csv('data/scientists.csv')
scientists
```

```python
scientists['Born']
```

```python
born_datetime = pd.to_datetime(scientists['Born'], 
                               format='%Y-%m-%d')
born_datetime.head(2)
```

```python
died_datetime = pd.to_datetime(scientists['Died'], 
                               format='%Y-%m-%d')
```

```python
scientists['born_dt'],scientists['died_dt'] = (born_datetime,
                                              died_datetime)
                                                
```

```python
scientists['age_days_dt'] = (scientists['died_dt'] - scientists['born_dt'] )
```
