```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10, 30)
y = np.sin(x)
plt.plot(x,y, '-ok')
plt.show()
```

```python
rnd = np.random.RandomState(0)
for marker in ['o','.',',','x','+', 'v','^','>','<', 's','d']:
    plt.plot(rnd.rand(5), rnd.rand(5), marker, label="marker='{0}'".format(marker))
    plt.legend()
    plt.xlim(0.0, 1.8)
```

```python
iris = pd.read_csv('data/iris_dataset.csv')
```

```python
iris.head(2)
```

```python
plt.scatter(iris.petal_length, iris.petal_width);
```

```python
set(iris.species)
```

```python
def check(x):
    if x == 'setosa':
        return 0
    elif x == 'versicolor':
        return 1
    else:
        return 2
```

```python
iris['species_'] = iris['species'].map(check)
```

```python
iris['species_color'] = iris['species'].map(
                            {'setosa':0, 'versicolor':1,'virginica':2})
```

```python
iris.iloc[100:110, :]
```

```python
plt.scatter(iris.petal_length, iris.petal_width, c = iris.species_color)
```

```python
plt.scatter(iris.petal_length, iris.petal_width, 
            c = iris.species_color,
            s = 50*iris.petal_length*iris.petal_width,
            alpha = 0.4)
```
