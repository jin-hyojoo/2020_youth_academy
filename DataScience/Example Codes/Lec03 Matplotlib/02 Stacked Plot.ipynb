```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5,6])
Apr =[5,7,6,8,7,9]
May = [0,4,3,7,8,9]
June = [6,7,4,5,6,8]
labels =['April', 'May', 'June']
fig, ax = plt.subplots()
ax.stackplot(x, Apr, May,June, labels =labels )
ax.legend(loc=2)
plt.xlabel('defect reason code')
plt.ylabel('number of defect')
plt.title('Product Defects -Q1 FY2019')
plt.show()
```

#### TableChart 

```python
rows =['2011','2012','2013', '2014','2015']
columns =['7Ah','35Ah', '40Ah','135Ah', '150Ah']
data=[[75, 144,114,102,108],
      [90, 126,102,84,126],
      [96, 114,75,105,135],
      [105, 90,150,90,75],
      [90, 75,135,75,90],
     ]
values = np.arange(0,600, 100)
colors = plt.cm.OrRd(np.linspace(0, 0.5, len(rows)))
n_rows = len(data)
index = np.arange(len(columns))+0.3
bar_width = 0.5
y_offset = np.zeros(len(columns))
fig, ax = plt.subplots()
cell_text = []
for row in range(n_rows):
    plot = plt.bar(index, data[row], bar_width, bottom=y_offset, color=colors[row])
    y_offset = y_offset+ data[row]
    cell_text.append(['%1.1f' %(x) for x in y_offset])
    i = 0
    for rect in plot:
        ax.text(rect.get_x()+ rect.get_width()/2, y_offset[i], '%d' % 
                int(y_offset[i]), ha='center', va='bottom')
        i = i +1

the_table = plt.table(cellText=cell_text, rowLabels=rows, 
                     rowColours=colors, colLabels=columns, loc='bottom')
plt.ylabel("Units Sold")
plt.title('Number of Battery Sold/year')
plt.xticks([])
plt.show()
```
