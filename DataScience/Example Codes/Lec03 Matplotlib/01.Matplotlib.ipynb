```python
x = np.linspace(0, 10, 100)
```

```python
figure = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
plt.show()
```

```python
figure.savefig('my_figure.png')
```

```python
figure.canvas.get_supported_filetypes()
```

```python
plt.subplot(2,2,1)
plt.plot(x, np.sin(x), '-')
plt.subplot(2,2,2)
plt.plot(x, np.cos(x), '--')
plt.subplot(2,2,3)
plt.plot(x, np.tan(x), '-')
plt.show()
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
ax[0].plot(x, np.tan(x))
```
